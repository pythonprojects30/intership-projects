```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('BitcoinPrice.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
data.drop(['Date'],1,inplace=True)
```

```python
data.head()
```

```python
#finding predcition for nect 30 days
```

```python
p_days=30                                                   #assigning no of days (30) to a variable and making a new col
data['prediction']=data[['Price']].shift(-p_days)           #prediction to get the next 30 day 
```

```python
data.head()
```

```python
data.tail
```

```python
#NaN stands for null values ,so droping it
```

```python
#separate x and y
```

```python
#creating x array
```

```python
x=np.array(data.drop(['prediction'],1))
```

```python
x=x[:len(data)-p_days]
```

```python
x
```

```python
#creating y array
```

```python
y=np.array(data['prediction'])
```

```python
y=y[: -p_days]
```

```python
y
```

```python
#split into train and test data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
p_day_array=np.array(data.drop(['prediction'],1))[-p_days :]
```

```python
p_day_array
```

```python
#estimation
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
Rf=RandomForestRegressor(n_estimators=1000,random_state=1)
```

```python
Rf.fit(x_train,y_train)
```

```python
print('Random Forest Accuracy: {:.2f}%'.format(Rf.score(x_test,y_test)*100))
```

```python
#PREDICTION
```

```python
Rf_pred=Rf.predict(x_test)
```
